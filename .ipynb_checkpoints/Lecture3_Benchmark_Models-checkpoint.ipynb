## 2) Ein simples Trend-Saisonalität Modell

## 2.1) Lineare Einfachregression

### 2.1.1) Berechnung der Trendgerade

Wie bei allen Prognoseaufgaben separiert man zunächst Daten fürs Training und für die Modellvalidierung

```python
X = np.arange(1,145,1) # tabelliere x-Werte: Start, Ende (ausgeschlossen), Schrittweite 
print(X)
```

```python
XTrain=X[0:108] # 9 Jahre Training 
XVal=X[108:] # 3 Jahre Validierung 
```

```python
PassengersTrain=Passengers[0:108] # 9 Jahre Training
PassengersVal=Passengers[108:] # 3 Jahre Validierung
              
#print(Passengers) 
#print(PassengersTrain)
#print(PassengersVal)                                    
```

Jetzt berechnen wir ein erstes lineares Modell, also den zeitlichen Trend.

Das lineare Modell ist gegeben durch:  

$f(t)= a+ b\cdot t$ hier heißt das $f(t) = intercept + slope\cdot t$ -  

ist also eine Funktion der Zeit $t$.  

Die Funktion "linregress" aus dem scipy stats-Paket berechnet einen Least Squares Fit (optimale Parameter durch Minimieren einer quadratischen Fehlerfunktion auf der Trainigsmenge).  
Berechnet werden:  
Steigung = slope, Achsenabschnitt = intercept sowie der Korrelationskoeffizient = r_value und als statistische Eigenschaften der Schätzung der Signifikanzwert = p_value und der Standardfehler der Schätzung = std_err für den Steigungskoeffizienten.

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(XTrain,PassengersTrain)
```

Jetzt ist es mögliche eine erste Prognose abzugeben indem man die nächsten Werte für t in die Transfer-Funktion einsetzt...   
Auswerten auf Trainings und Validierungsmenge:

```python
# definiere eine Funktion für die Trendgerade: 

def LinPredict(x):
    return intercept + slope*x
```

```python
PredictLinTrain = LinPredict(XTrain)
PredictLinVal = LinPredict(XVal) 
print(PredictLinVal)
```

```python
plt.plot(XTrain,PassengersTrain, '-', label='original data'); # Plot mit Legende für die Daten
plt.plot(XTrain, PredictLinTrain, 'r', label='Linear Model'); # Plot mit Legende für die Daten
plt.ylabel('Number of Passengers'); 
plt.xlabel('Time Step');
plt.legend(); # Legende einfügen
plt.title('Linear fit on Training set'); #Titel der Graphik 
```

```python
plt.plot(XVal,PassengersVal, '-', label='Original Data');
plt.plot(XVal,PredictLinVal, 'r', label='Linear Model');
plt.ylabel('Number of Passengers'); 
plt.xlabel('Time Step');
plt.legend(); # Legende einfügen
plt.title('Linear fit on Validation set'); #Titel der Graphik 
```

### 2.1.2) Bewertung der Prognose mit Hilfe quantitativer Maße

Um zu sagen wie gut die Prognose ist braucht man ein Fehlermaß z.B. den Root Mean Square error (RMSE).  
Man kann sich entweder selbst eine Funktion definieren, die den RMSE $\frac{1}{n}\sum_{i=1}^{n}(out_i - tar_i)^2$ auswertet, oder eingebaute Funktionalitäten benutzen.

```python
from sklearn.metrics import mean_squared_error 
from math import sqrt
rmseLMVal = sqrt(mean_squared_error(PassengersVal, PredictLinVal)) 
rmseLMVal
```

### 2.1.3) Analyse der Residuen

Eine weitere Methode die Modellgüte zu evaluieren ist eine Auswertung der Residuen (Output - Target).

```python
ResLMVal = PredictLinVal - PassengersVal 
ResLMVal.head()
```

Mittelwert der Residuen:
    
$\rightarrow$ ist der Mittelwert negativ, so unterschätzt das Modell systematisch, ist er positiv, so wird überschätzt!


```python
meanResLMVal=ResLMVal.mean() 
meanResLMVal
```

```python
def f(t):
     return meanResLMVal+(t-t) 
# ganz blöder Trick um eine konstante Funktion zu haben für den Plot, die für Listen funktioniert...
```

```python
f(XVal)
```

```python
plt.plot(XVal, ResLMVal, '-', label='Residuals');
plt.plot(XVal,f(XVal), 'r',label='Mean of Residuals');
plt.ylabel('Number of Passengers'); 
plt.xlabel('Time Step');
plt.legend();
plt.title('Residuals from Linear Fit on Validation set'); #Titel der Graphik
```

...der mittlere Residualfehler ist stark negativ  

$\rightarrow$ das Model liegt systematisch unter den tatsächlichen Werten!  
Man erkennt aber auch ganz deutlich, daß die Fehler mit fortschreitender Zeit zunehmen, da die Varianz in den Daten wächst.  
Ein linearer Trend ist hier wahrscheinlich nicht besonders geeignet!

## 2.2) Exponentialer Trend

### 2.2.1) Schätzen des Modelles

```python
# importiere den benötigten scipy optimizer zum fitten von beliebigen Kurven
from scipy.optimize import curve_fit 

# definiere die Funktion, die an die Daten angepaßt werden soll
def func(x, a, b, c):
    return a * np.exp(-b * x) + c
# führe den Fit durch!
# ausgegeben werden die Fit-Parameter und die Kovarianzmatrix der Parameter
popt, pcov = curve_fit(func, XTrain,PassengersTrain)
print(popt)
```

### 2.2.2) Auswertung des Modelles

Auswerten des Modelles auf Training und Validierungsmenge

```python
PredictExpTrain=func(XTrain,*popt) 
PredictExpVal=func(XVal,*popt)
```

graphische Darstellung der Daten (Training und Validierung) zusammen mit den beiden Trend-Modellen

```python
plt.plot(XTrain,PassengersTrain, '-', label='Original Data - Training');
plt.plot(XTrain, PredictLinTrain, 'r', label='Linear Model');
plt.plot(XTrain,PredictExpTrain , 'g', label='Exponential Model'); 

plt.plot(XVal,PassengersVal, '-', label='Original Data - Validation');
plt.plot(XVal,PredictLinVal, 'r', label='Linear Model');
plt.plot(XVal,PredictExpVal, 'g', label='Exponential Model');
plt.ylabel('Number of Passengers'); 
plt.xlabel('Time Step');
plt.legend();
plt.title('Comparison of linear and exponential trend'); # Titel der Graphik
```

```python
ResExpModelVal = PredictExpVal - PassengersVal # Berechnung der Residuen
meanResExpModelVal = ResExpModelVal.mean() # Mittelwert der Residuen
rmseExpVal = sqrt(mean_squared_error(PassengersVal, PredictExpVal)) # RMSE des exponentiellen Modelles

print('Mittelwert Residuen lineares Modell', meanResLMVal) 
print('RMSE lineares Modell', rmseLMVal)
print('Mittelwert Residuen exponentielles Modell',meanResExpModelVal)
print('RMSE exponentielles Modell',rmseExpVal)
```

## 2.3) Ein Modell für die Residuen

Der nächste Schritt ist jetzt zu versuchen den Restfehler durch ein geeignetes Modell zu erklären.  
Dazu berechnet man zunächst trendbereinigte Daten (Trainingsmenge, Validierungsmenge)...  
Das lineare Modell nehmen wir als Indikator für den Trend.      

```python
ResLinTrain=PassengersTrain-PredictLinTrain 
ResLinVal=PassengersVal-PredictLinVal
```

```python
plt.plot(XTrain,ResLinTrain, '-', label='Detrended data')
plt.ylabel('Number of Passengers'); 
plt.xlabel('Time Step');
plt.legend();
plt.title('Detrended Data on Training set'); #Titel der Graphik
```

ein reines saisonales Modell ist hier vielleicht nicht so gut geeignet, weil die Residuen ein klares Wachstumssignal anzeigen.  
$\rightarrow$ Exponential Smoothing oder autoregressives Modell

```python
from scipy import stats #statistics tool das viele build-in Modelltypen enthält 
import statsmodels.api as sm
import statsmodels.formula.api as smf 
```

```python
fig = plt.figure(figsize=(12,8)) 
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ResLinTrain, lags=20, ax=ax1) #plot der Autokorrelationsfunktion bis zum Lag 20
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(ResLinTrain, lags=20, ax=ax2) #plot der partiellen Autokorrelationsfkt. bis zum Lag 20
```

```python
ResLinTrain = np.array(ResLinTrain) 
ResLinVal = np.array(ResLinVal)
PassengersVal = np.array(PassengersVal) 
type(ResLinTrain)
```

```python
arma_mod20 = sm.tsa.ARMA(ResLinTrain, (2,0)).fit() #z.B. ein ARMA(2,0)-Modell an die Daten fitten
```

```python
print(arma_mod20.params) #geschätzte Parameter ausgeben
```

Will man mit Hilfe eines solchen Modelles Prognosen machen ist es günstig die Daten als Pandas Time-Series Objekt zu formatieren. Das erlaubt viele zusätzliche Funktionalitäten. Zeitstempel werden z.B. automatisch erkannt. Es ist möglich auf bestimmte Zeitauflösung zu aggregieren, Zeitreihen mit unterscheidlichen Anfangs- und Endpunkten automatisch richtig zusammenzusetzen, ...
                            

```python
ResLinTrain.shape
dates = sm.tsa.datetools.dates_from_range('1949m1', length=108) #definiere Datumsbereich
TrainDates = sm.tsa.datetools.dates_from_range('1949m1', length=108)
ValDates=sm.tsa.datetools.dates_from_range('1958m1', length=36)
y = pd.Series(ResLinTrain, index=TrainDates) #Pandas Time Series mit Zeitstempel im korrekten Zeitfenster
arma_mod = sm.tsa.ARMA(y, order=(3,2)) #ARMA(3,2) Modell der trendbereinigten Daten 
arma_res = arma_mod.fit(trend='nc', disp=-1) #Option trend='c' : Konstante C im ARMA-Modell berücksichtigt ('nc' sonst)
                                     
```

```python
TrainDates;
```

```python
PredictARMATrain = arma_res.predict(start='1949-01-31',end='1957-12-31') #Vorhersage für die Validierungsmenge
ResLinValSeries = pd.Series(ResLinVal, index=ValDates) #trendbereinigte Daten als Zeitreihe
                               
```

```python
plt.plot(y, '-', label='Detrended Original Data') 
plt.plot(PredictARMATrain, 'r', label='ARMA Model')
plt.plot(ResLinValSeries, '-', label='Detrended Original Data') 
plt.legend()
plt.title('Prognosis from ARMA Model on Validation set') #Titel der Graphik
plt.show()
```

```python
PredictLinTrainSeries = pd.Series(PredictLinTrain, index=TrainDates) 
PassengersTrainSeries = pd.Series(PassengersTrain, index=TrainDates)
```

jetzt kann man ein additiv zusamengesetztes Modell auswerten mit xpred(t) = Linpred(t) + ARMApred(t)

```python
AdditiveModel = PredictLinTrainSeries + PredictARMATrain #zusammengestztes Modell 
#print(AdditiveModel);
```

```python
plt.plot(PredictLinTrainSeries+y, 'b', label='Original Data')
plt.plot(AdditiveModel, 'r', label='ARMA + Linear Model') 
plt.legend()
plt.title('Prognosis from ARMA + Linear Model') #Titel der Graphik
```

Auswertung des Ergebnisses mit Hilfe des RMSE

```python
rmse = sqrt(mean_squared_error(PassengersTrain, PredictLinTrain))
rmseARMAplusLin = sqrt(mean_squared_error(PredictLinTrainSeries+y,AdditiveModel))
print(rmse)
print(rmseARMAplusLin)
```
