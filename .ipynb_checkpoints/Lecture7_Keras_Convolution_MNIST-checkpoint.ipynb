## Vordefinierte Funktion um den Fehler während des Trainings zu beobachten

```python
# updatable plot
# a minimal example (sort of)
from IPython.display import clear_output
class PlotLosses(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.i = 0
        self.x = []
        self.losses = []
        self.val_losses = []
        
        self.fig = plt.figure()
        
        self.logs = []

    def on_epoch_end(self, epoch, logs={}):
        
        self.logs.append(logs)
        self.x.append(self.i)
        self.losses.append(logs.get('loss'))
        self.val_losses.append(logs.get('val_loss'))
        self.i += 1
        
        clear_output(wait=True)
        plt.plot(self.x, self.losses, label="loss")
        plt.plot(self.x, self.val_losses, label="val_loss")
        plt.legend()
        plt.show();
        
plot_losses = PlotLosses()
```

## Import der MNIST Daten in kompakter Form - Quelle Keras datasets

```python
#reshape
x_train_unnorm = x_train_orig.reshape(60000,28,28,1)
x_test_unnorm = x_test_orig.reshape(10000,28,28,1)
```

```python
print(x_train_unnorm[0].max())
print(x_test_unnorm[0].max())
```

```python
x_train=x_train_unnorm/255
x_test=x_test_unnorm/255
```

```python
print(x_train[0].max())
print(x_test[0].max())
```

Das Target muß noch OneHot-kodiert werden

```python
#one-hot encode target column
y_train = to_categorical(y_train_orig)
y_test = to_categorical(y_test_orig)
y_train[0]
```

## erstes Convolution-Netz zur Klassifikation

```python
#create model
modelCNN_MNIST = Sequential()
#add model layers
modelCNN_MNIST.add(Conv2D(64,                       #Anzahl der Filter
                          kernel_size=(3,3),        #Dimensionen des Fensters
                          strides=(1,1),            #Größe der Schritte
                          padding='same',           #Padding: same=dim(in)=dim(out), valid=no Padding
                          activation='relu',        #Aktivierungsfunktion
                          input_shape=(28,28,1)))   #Dimension des Inputs

modelCNN_MNIST.add(Conv2D(32, kernel_size=(3,3), 
                          strides=(1,1),            #Größe der Schritte
                          padding='same',           #Padding: same=dim(in)=dim(out), valid=no Padding
                          activation='relu',        #Aktivierungsfunktion
                          input_shape=(28,28,1)))

modelCNN_MNIST.add(Flatten())                       #Feature-Vektor erzeugen
modelCNN_MNIST.add(Dense(10, activation='softmax')) #dicht verschaltete Schicht - lernt Klassifikation!

from keras.utils import plot_model
plot_model(modelCNN_MNIST, to_file='modelCNN_MNIST.png')

#compile model using accuracy to measure model performance

#myAdam=keras.optimizers.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, amsgrad=False)
#myRMSprop=keras.optimizers.RMSprop(learning_rate=0.001, rho=0.9)


#Stochastic gradient descent
sgd = SGD(lr=0.001, decay=1e-6, momentum=0.9, nesterov=True) #Parameter beim SGD Lernrate, Momentum, ...

modelCNN_MNIST.compile(
    optimizer=sgd, 
    loss='categorical_crossentropy', 
    metrics=['accuracy'])

#train the model
historyCNN_MNIST=modelCNN_MNIST.fit(x_train, y_train, 
          epochs=5,
          batch_size=128,
          verbose=1,
          validation_data=(x_test, y_test),  
          callbacks=[plot_losses])
```

```python
from keras.utils import plot_model
plot_model(modelCNN_MNIST, to_file='modelCNN_MNIST.png')
```

```python
print(modelCNN_MNIST.summary())
```

```python
#Prediction für das erste Bild der Testmenge
resultCNN_MNIST=modelCNN_MNIST.predict(x_test[:1])
print(resultCNN_MNIST)
```

```python
print ('output')
print (np.argmax(resultCNN_MNIST))
 
print ('target') 
print (np.argmax(y_test[:1]))
```

```python
batch_size = 128
num_classes = 10
epochs = 10

modelCNN2 = Sequential()
modelCNN2.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=(28,28,1)))

modelCNN2.add(Conv2D(64, (3, 3), 
                 activation='relu'))

modelCNN2.add(MaxPooling2D(pool_size=(2, 2)))
modelCNN2.add(Dropout(0.25))
modelCNN2.add(Flatten())
modelCNN2.add(Dense(128, activation='relu'))
modelCNN2.add(Dropout(0.5))
modelCNN2.add(Dense(num_classes, activation='softmax'))

modelCNN2.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])

historyCNN2=modelCNN2.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test),
          callbacks=[plot_losses])
score = modelCNN2.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

## geht auch mit Hilfe der history - allerdings erst, wenn das Training abgeschlossen ist

```python
# Plotte Training & Validation accuracy (Hit Rate)
plt.plot(historyCNN2.history['accuracy'])
plt.plot(historyCNN2.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plotte Training & Validation loss (hängt von der Loss-Funktion beim Training ab)
plt.plot(historyCNN2.history['loss'])
plt.plot(historyCNN2.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
